# District Summary


```python
# Calculate the total number of schools
total_school = school_data['school_name'].count()
total_school


```

```python
# Calculate the total number of students
student_number = student_data["Student ID"].count()
student_number
```

```python
# Calculate the total budget
a = school_data['budget'].sum()
b = str("$") + str(a)
b
```

```python
# Average Math Score
math_score_mean = round(student_data['math_score'].mean(),6)
math_score_mean
```

```python
# Average Reading Score
Average_reading_score = round(student_data['reading_score'].mean(),5)
Average_reading_score
```

```python
#  Passing Math percentage
passing_studentmath = len(student_data[student_data['math_score']>=70])
percentageMath = round(passing_studentmath/student_number * 100,6)
percentageMath

```

```python
#  Passing Reading percentage
passing_studentreading = len(student_data[student_data['reading_score']>=70])
percentageReading = round(passing_studentreading/student_number * 100,6)
percentageReading
```

```python
# Passing math and reading percentage
passing_student = len(student_data[(student_data.reading_score >= 70) & (student_data.math_score >= 70)])
Percentage_passingstudent = round(passing_student/student_number * 100, 6)
Percentage_passingstudent
```

```python
# initialize data of lists.
data = {'Total Schools': [total_school],
        'Total Students': [student_number],
       'Total Budget':[math_score_mean],
       'Average Math Score':[Average_reading_score],
       '% Passing Math':[percentageMath],
       '% Passing Reading':[percentageReading],
       '% Overall Passing':[Percentage_passingstudent],
       'Total Budget':[b]}


# Create DataFrame
df = pd.DataFrame(data)
  
# Print the output.
df
```

# School Summary

```python
# School Name
School_Name = student_data['school_name'].unique()
School_Name


```

```python
# School Type
School_Type = school_data.groupby('school_name')['type'].sum()
School_Type
```

```python
# Total student
Total_student = school_data.groupby('school_name')['size'].sum()
Total_student
```

```python
# Total Student Budget
Total_schoolBudget = school_data.groupby('school_name')['budget'].sum()
Total_schoolBudget
```

```python
#Per Student Budget
Per_studentBudget = Total_schoolBudget/Total_student
Per_studentBudget
```

```python
# Average Math Score
total_mathscore = student_data.groupby(['school_name'])['math_score'].agg('sum')
# total_mathscore
total_student_number = student_data.groupby(['school_name'])['math_score'].agg('count')
# Average Math Score for each school
AverageMathScore = total_mathscore/total_student_number
AverageMathScore
```

```python
# Average Reading Score
total_readingscore = student_data.groupby(['school_name'])['reading_score'].agg('sum')

# Average reading Score for each school
AverageReadingScore = total_readingscore/total_student_number
AverageReadingScore
```

```python
# Calculate total pass math students
total_mathscore = student_data[student_data['math_score']>70].groupby(['school_name'])['math_score'].agg('count')
percentagemath = total_mathscore/Total_student*100
percentagemath

```

```python
# Calculate total pass reading students
total_readingscore = student_data[student_data['reading_score']>70].groupby(['school_name'])['reading_score'].agg('count')
percentagereading = total_readingscore/Total_student*100
percentagereading
```

```python
# Overall Passing (The percantage of students that passed math and reading)
total_pass = student_data[
    (student_data['reading_score']>70) 
    & (student_data['math_score']>70)].groupby(['school_name'])['reading_score'].agg('count')

total_mathscore = total_pass / Total_student * 100
total_mathscore
                               
    
    

                               
```
